## Exercise 1

```python
# Import cell 
import numpy as np
import matplotlib.pyplot as plt
```

### Ex 1.a

```python
# A) Rejection Method
R = 1
N = 2000
points_in_circle = list()
points_out_circle = list()


while len(points_in_circle) <= N:
    accept = False
    x = np.random.uniform(-1, 1)
    y = np.random.uniform(-1, 1)
    point = np.array([x, y])
    
    if np.square(x) + np.square(y) <= 1:
        accept = True
    
    if accept: 
        points_in_circle.append(point)
    else: 
        points_out_circle.append(point)

        
points_in_circle = np.array(points_in_circle)
points_out_circle = np.array(points_out_circle)


# Plot
plt.figure(figsize = [6 ,6])
plt.scatter(points_in_circle[:,0], points_in_circle[:,1], s = 3, edgecolors = 'g')
plt.scatter(points_out_circle[:,0], points_out_circle[:,1], s = 3, edgecolors = 'r')
plt.xlabel('x', fontsize = 13, fontstyle = "italic")
plt.ylabel('y', fontsize = 13, fontstyle = "italic")
plt.title("Rejection Method", fontsize = 15, fontstyle = "italic")
plt.grid()
plt. tight_layout()
plt.show ()
```

### Ex 1.b

```python
# Accept all points 
N = 2000

r = np.random.uniform(0, 1, N)
theta = np.random.uniform(0, 1, N) * 2 * np.pi

x_acc = r * np.cos(theta)
y_acc = r * np.sin(theta)


plt.figure(figsize = [6 ,6])
plt.scatter(x_acc, y_acc, s = 1, facecolors = 'none', edgecolors = 'g')
plt.xlabel('x', fontsize = 13, fontstyle = "italic")
plt.ylabel('y', fontsize = 13, fontstyle = "italic")
plt.title("Accepting all points with the help of R and Theta" , fontsize = 15, fontstyle = "italic")
plt.grid()
plt. tight_layout ()
plt.show ()
```

 In this second exercise, the distribution is **not** uniform over the circle of radius R. Instead most of the points are concentrated close to the origin $(0,0)$. The motivation behind this wrong result is that the points are uniformly generated with respect to the radius $r$.

### Ex 1.c

```python
# C. Inverse distribution Method

R = 1
N = 2000
C = 1/(np.pi * R**2)


theta = np.random.uniform(0, 1, N) * 2 * np.pi

u = np.random.uniform(0, 1, N)
r = np.sqrt(u)


x = r * np.cos(theta)
y = r * np.sin(theta)


plt.figure(figsize = [6 ,6])
plt.scatter(x, y, s = 1, facecolors = 'none', edgecolors = 'g')
plt.xlabel('x', fontsize = 13, fontstyle = "italic")
plt.ylabel('y', fontsize = 13, fontstyle = "italic")
plt.title("Inverse Distribution Method", fontsize = 15, fontstyle = "italic")
plt.grid()
plt. tight_layout ()
plt.show ()
```

Here there are reported the calculus to compute the value of $C$.

Starting from the distribution of $r$ that is given by text: $p(r) = C2\pi r $. It must be integrated:
$$\begin{aligned}
 & F (r) = \int_{0}^{r} C 2 \pi x\ dx  = \\
 \\
 & C2\pi \int_{0}^{r}x\ dx = \\
 \\
 & C 2\pi \frac{r^2}{2} = C\pi r^2
\end{aligned}$$

And then compute $F(r = R) = 1$, from where it is obtained $C \pi R^2 = 1 \Rightarrow C = \frac{1}{R^2 \pi}\ \big|_{R = 1} = \frac{1}{\pi}$. After that the constant $C$ is computed we can compute the inverse of the cumulative distribution. In the following way: 

$$\begin{aligned}
 & F(r) = C \pi r^2  = y\\
 \\
 &r^2 = \underbrace{\frac{y}{C \pi}}_{C = \frac{1}{\pi}} = \\
 \\
 & r = \sqrt{y}
\end{aligned}$$

```python
# Comparison
fig, axs = plt.subplots(1, 2, figsize = (12,6))
axs[0].scatter(x, y, s = 1, facecolors = 'none', edgecolors = 'g')
axs[0].set_xlabel('x', fontsize = 13, fontstyle = "italic")
axs[0].set_ylabel('y', fontsize = 13, fontstyle = "italic")
axs[0].set_title("Inverse Distribution Method", fontsize = 15, fontstyle = "italic")
axs[0].grid()


axs[1].scatter(x_acc, y_acc, s = 1, facecolors = 'none', edgecolors = 'g')
axs[1].set_xlabel('x', fontsize = 13, fontstyle = "italic")
axs[1].set_ylabel('y', fontsize = 13, fontstyle = "italic")
axs[1].set_title("Accepting all points with the help of R and Theta", fontsize = 15, fontstyle = "italic")
axs[1].grid()

plt.show ()
```

As it is possible to observe from the above graphs: the inverse distribution method is well uniform distributed, while the method used in point B is not. In the latter case, the points seem to concentrate close to the origin. This result is obtained since the points are uniformly distributed according to the radius and not the area of the circle.

## Exercise 2

From the text of the exercise we know that $x$ follows the following distribution: $f(x) = C x^{-\frac{3}{2}}$. To compute the value of $C$ we must have to integrate over the support $[1, \infty)$:

$$\begin{aligned}
 & \int_{1}^{\infty} C x^{-\frac{3}{2}}\ dx  = 1\\
 \\
 & C \int_{1}^{\infty}x^{-\frac{3}{2}}\ dx = 1\\
 \\
 & C \left[\frac{x^{-\frac{1}{2}}}{-\frac{1}{2}}\right]_{1}^{\infty} = 1\\
 \\
 & 2C = 1 \Rightarrow C = \frac{1}{2}
\end{aligned}$$
Now we can find the cumulative distribution in the following way:

$$\begin{aligned}
 & F(x) = \int_{1}^{x} \frac{1}{2} r^{-\frac{3}{2}}\ dr  = \\
 \\
 & \frac{1}{2} \int_{1}^{x}r^{-\frac{3}{2}}\ dr = \\
 \\
 & \frac{1}{2} \left[\frac{r^{-\frac{1}{2}}}{-\frac{1}{2}}\right]_{1}^{x} = \\
 \\
 & \left[-r^{-\frac{1}{2}}\right]_{1}^{x} = \left[-x^{-\frac{1}{2}} + 1 \right] 
\end{aligned}$$
Now that we discover that $F(x) = 1 -x^{-\frac{1}{2}} $ we put it equal to $y $ and we compute the inverse.
$$\begin{aligned}
 & 1 -x^{-\frac{1}{2}} = y \\
 \\
 & 1-y = x^{-\frac{1}{2}} \\
 \\
 & \frac{1}{(1-y)^2} = x \\
\end{aligned}$$

So the inverse is $F^{-1}(y) = \frac{1}{(1-y)^2}$

#### Linear binning

```python
A = 200000
B = 190000
N = 10000
C = 1/2

def inverse_CDF(y):
    return 1/np.square(1-y)


y = np.random.uniform(0, B, N)
scaled_y = y/A

x = inverse_CDF(scaled_y)

START_VALUE = 1
STOP_VALUE = 1000
NUM_OF_BINS = 30 

lin_bins = np.linspace(start = START_VALUE, stop= STOP_VALUE, num = NUM_OF_BINS)

x_s = np.arange(1, 1000)

fig, axs = plt.subplots(2, 2, figsize = (15,13))

axs[0, 0].hist(x, facecolor = 'cyan', edgecolor = 'b', bins = lin_bins, linewidth = 1, density = True)
axs[0, 0].set_title(f'Histogram-Linear coordinates', fontstyle = 'italic')
axs[0, 0].set_xlabel ('Mass', fontsize = 10)
axs[0, 0].set_ylabel ('PMF', fontsize = 10)
axs[0, 0].set_xlim(START_VALUE, STOP_VALUE)
axs[0 ,0].plot(x_s ** (-3/2))
axs[0, 0].grid()

axs[0, 1].hist(x, facecolor = 'cyan', edgecolor = 'b', bins = lin_bins, linewidth = 1, density = True)
axs[0, 1].set_title(f'Histogram- Log coordinates', fontstyle = 'italic')
axs[0, 1].set_xlabel ('Mass', fontsize = 10)
axs[0, 1].set_ylabel ('PMF', fontsize = 10)
axs[0, 1].set_xlim(START_VALUE, STOP_VALUE)
axs[0 ,1].plot(5/7 *x_s ** (-3/2))
axs[0, 1].set_yscale('log')
axs[0, 1].set_xscale('log')
axs[0, 1].grid()

# Compute midpoints
y, edges = np.histogram(x, lin_bins, density = True)
centers = 0.5*(edges[1:] + edges[:-1])


axs[1, 0].plot(x_s ** (-3/2))
axs[1, 0].plot(centers, y, "*")
axs[1, 0].set_title(f'Midpoints- Linear coordinates', fontstyle = 'italic')
axs[1, 0].set_xlabel ('Mass', fontsize = 10)
axs[1, 0].set_ylabel ('PMF', fontsize = 10)
axs[1, 0].set_xlim(START_VALUE, STOP_VALUE)
axs[1, 0].grid()



axs[1, 1].plot(C *x_s ** (-3/2))
axs[1, 1].plot(centers, y, "*")
axs[1, 1].set_title(f'Midpoints- Log coordinates', fontstyle = 'italic')
axs[1, 1].set_xlabel ('Mass', fontsize = 10)
axs[1, 1].set_ylabel ('PMF', fontsize = 10)
axs[1, 1].set_yscale('log')
axs[1, 1].set_xscale('log')
axs[1, 1].set_xlim(START_VALUE, STOP_VALUE)
axs[1, 1].grid()


plt.show()
```

#### Logarithmic binning

```python
log_bins = np.logspace(start = np.log10(1), stop= np.log10(1000), num = NUM_OF_BINS)

fig, axs = plt.subplots(2, 2, figsize = (15,13))

axs[0, 0].hist(x, facecolor = 'cyan', edgecolor = 'b', bins = log_bins, linewidth = 1, density = True)
axs[0, 0].set_title(f'Histogram-Linear coordinates', fontstyle = 'italic')
axs[0, 0].set_xlabel ('Mass', fontsize = 10)
axs[0, 0].set_ylabel ('PMF', fontsize = 10)
axs[0, 0].set_xlim(START_VALUE, STOP_VALUE)
axs[0 ,0].plot(x_s ** (-3/2))
axs[0, 0].grid()

axs[0, 1].hist(x, facecolor = 'cyan', edgecolor = 'b', bins = log_bins, linewidth = 1, density = True)
axs[0, 1].set_title(f'Histogram- Log coordinates', fontstyle = 'italic')
axs[0, 1].set_xlabel ('Mass', fontsize = 10)
axs[0, 1].set_ylabel ('PMF', fontsize = 10)
axs[0, 1].set_xlim(START_VALUE, STOP_VALUE)
axs[0 ,1].plot(5/7 *x_s ** (-3/2))
axs[0, 1].set_yscale('log')
axs[0, 1].set_xscale('log')
axs[0, 1].grid()

# Compute midpoints
y, edges = np.histogram(x, log_bins, density = True)
centers = 0.5*(edges[1:] + edges[:-1])


axs[1, 0].plot(x_s ** (-3/2))
axs[1, 0].plot(centers, y, "*")
axs[1, 0].set_title(f'Midpoints- Linear coordinates', fontstyle = 'italic')
axs[1, 0].set_xlabel ('Mass', fontsize = 10)
axs[1, 0].set_ylabel ('PMF', fontsize = 10)
axs[1, 0].set_xlim(START_VALUE, STOP_VALUE)
axs[1, 0].grid()



axs[1, 1].plot( 1/2 *x_s ** (-3/2))
axs[1, 1].plot(centers, y, "*")
axs[1, 1].set_title(f'Midpoints- Log coordinates', fontstyle = 'italic')
axs[1, 1].set_xlabel ('Mass', fontsize = 10)
axs[1, 1].set_ylabel ('PMF', fontsize = 10)
axs[1, 1].set_yscale('log')
axs[1, 1].set_xscale('log')
axs[1, 1].set_xlim(START_VALUE, STOP_VALUE)
axs[1, 1].grid()


plt.show()
```

The best plot to better visualize the distribution is midpoints in log coordinates with log binning for these two reasons:
* Midpoints create a straight line and the slope of the straight line corresponds to the power law exponent.
* The log-log scale is necessary since it emphasizes the upper tail region. 

#### 2.e

The rejection method would be slower since sometimes points will be rejected. If we think of these simulation methods as processes, the one that is more analogous to how realizations $x \in X$ emerge in a natural stochastic process is the **inverse distribution method**. It is the only method where each point (realization) belongs to the uniform distribution in the circle.

In the first method (1.a), not every point belongs to the uniform distribution inside the circle of radius $R$. Instead, the second method (1.b) is not a realization of a uniform distribution.
