```python
# Import cell
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Utils: set all the constants
N = 20 
MEAN = 30 
VAR = 20 
TAU = 1/VAR

A = 1
B = 1
C = 3
D = 1/100

BURN_IN = 20
NR_SIMULATION = 100000
```

```python
def normal(tau):
    den = N*tau+D
    mean = (C*D + N*tau*MEAN)/den
    std = np.sqrt(1 / den)
    return np.random.normal(mean, std)

def gamma(mu):
    k = A + N/2
    scale = B + 0.5*((N-1)*VAR + N*(MEAN-mu)**2)
    return np.random.gamma(k, 1/scale)

mus = np.zeros(NR_SIMULATION+1)
taus = np.zeros(NR_SIMULATION+1)

mus[0] = A/C
taus[0] = C

for j in range(NR_SIMULATION):
    mu = normal(taus[j])
    tau = gamma(mus[j])
    
    mus[j+1] = mu
    taus[j+1] = tau
```

```python
fig, axs = plt.subplots(1, 2, figsize=[10, 5])

X_AXIS = np.arange(NR_SIMULATION+1)

axs[0].plot(X_AXIS, mus, color= "darkorange")
axs[0].set_xlabel ('Iteration Steps', fontsize = 10)
axs[0].set_ylabel ('$\mu$', fontsize = 10)
axs[0].set_title(f'$\mu$ vs iteration steps', fontstyle = 'italic')
axs[0].grid()

axs[1].plot(X_AXIS[BURN_IN:], mus[BURN_IN:], color= "darkorange")
axs[1].set_xlabel ('Iteration Steps', fontsize = 10)
axs[1].set_ylabel ('$\mu$', fontsize = 10)
axs[1].set_title(f'$\mu$ vs iteration steps (burn-in)', fontstyle = 'italic')
axs[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=[10, 5])

X_AXIS = np.arange(NR_SIMULATION+1)

axs[0].plot(X_AXIS, taus, color= "lightblue")
axs[0].set_xlabel ('Iteration Steps', fontsize = 10)
axs[0].set_ylabel ('$\\tau$', fontsize = 10)
axs[0].set_title(f'$\\tau$ vs iteration steps', fontstyle = 'italic')
axs[0].grid()

axs[1].plot(X_AXIS[BURN_IN:], taus[BURN_IN:], color= "lightblue")
axs[1].set_xlabel ('Iteration Steps', fontsize = 10)
axs[1].set_ylabel ('$\\tau$', fontsize = 10)
axs[1].set_title(f'$\\tau$ vs iteration steps (burn-in)', fontstyle = 'italic')
axs[1].grid()

plt.show()
```
