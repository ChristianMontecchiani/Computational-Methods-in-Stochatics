```python
# Import cell

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

from numpy.random import poisson
from numpy.random import uniform
from numpy.random import exponential
```

```python
# Utils: contains constant values for both algorithms

T = 24 * 60 

U_LAMBDA = 1

N = 1000

D = {(0 * 60, 6 * 60): 0.2, (6 * 60, 8 * 60): 0.9,\
     (8 * 60, 12 * 60): 0.5, (12 * 60, 15 * 60): 0.6,\
     (15 * 60, 17 * 60): 1.0, (17 * 60, 19 * 60): 0.6,\
     (19 * 60, 24 * 60): 0.5
    }

MEAN_ARRIVALS = list()
CUM_ARRIVALS = list()
```

## Revised Method

```python
# Initialization 
tot_arrivals = np.zeros(T)
first_car = np.zeros(N)
x_0 = 0 

for j in range(N):
    x_i = x_0
    m = poisson(U_LAMBDA * T)
    first = True
    for i in range(m):
        u = uniform(0,1)
        exponent = 1/(m-i+1)
        x_i = x_i+(T-x_i)*(1-u**exponent)
        y = uniform(0, U_LAMBDA)
        for a,b in D.keys(): 
            if x_i>=a and x_i<b: 
                l = D[(a,b)]
                if  y<=l:
                    idx = int(x_i)
                    tot_arrivals[idx] += 1
                    if first:
                        first_car[j] = x_i
                        first = False
                continue
                    

x_axis = np.arange(T)
mean_arrivals = tot_arrivals/N

#Plot
fig, axs = plt.subplots(1, 2, figsize=[12, 5])

axs[0].plot(x_axis, mean_arrivals)
axs[0].set_xlim(0,T)
axs[0].set_xlabel ('Time in minutes', fontsize = 8)
axs[0].set_ylabel ('Avg number of cars', fontsize = 10)
axs[0].set_title(f'Avg number of cars', fontstyle = 'italic')
axs[0].grid()

# Cumulative 
cum_arrivals = np.cumsum(tot_arrivals)/1000
axs[1].plot(x_axis, cum_arrivals)
axs[1].set_xlim(0,T)
axs[1].set_xlabel ('Time in minutes', fontsize = 8)
axs[1].set_ylabel ('Cumulative number of cars', fontsize = 10)
axs[1].set_title(f'Cumulative number of cars', fontstyle = 'italic')
axs[1].grid()

MEAN_ARRIVALS.append(mean_arrivals)
CUM_ARRIVALS.append(cum_arrivals)
```

```python
print(f"Average number of cars that arrived in 24 hours: {int(np.sum(tot_arrivals)/N)}")
print(f"Average time when the first car arrived: {np.mean(first_car):.2f}")
```

## Thinning Approach

```python
t_0 = 0 
i = 1

tot_arrivals = np.zeros(T)
first_car = np.zeros(N)

for j in range(N):
    t_i = t_0
    first = True
    while t_i<=T:
        t_i += exponential(scale= 1/U_LAMBDA)
        u = uniform(0,1)
        for a,b in D.keys(): 
            if t_i >= a and t_i<b: 
                l = D[(a,b)]/U_LAMBDA
                if u <= l:
                    if first:
                        first_car[j] = t_i
                        first = False
                    idx = int(t_i)
                    tot_arrivals[idx] += 1
                continue
                        
        i += 1
        
mean_arrivals = tot_arrivals/N
x_axis = np.arange(T)

#Plot
fig, axs = plt.subplots(1, 2, figsize=[12, 5])

axs[0].plot(x_axis, mean_arrivals)
axs[0].set_xlim(0,T)
axs[0].set_xlabel ('Time in minutes', fontsize = 8)
axs[0].set_ylabel ('Avg number of cars', fontsize = 10)
axs[0].set_title(f'Avg number of cars', fontstyle = 'italic')
axs[0].grid()

# Cumulative 
cum_arrivals = np.cumsum(tot_arrivals)/1000
axs[1].plot(x_axis, cum_arrivals)
axs[1].set_xlim(0,T)
axs[1].set_xlabel ('Time in minutes', fontsize = 8)
axs[1].set_ylabel ('Cumulative number of cars', fontsize = 10)
axs[1].set_title(f'Cumulative number of cars', fontstyle = 'italic')
axs[1].grid()

MEAN_ARRIVALS.append(mean_arrivals)
CUM_ARRIVALS.append(cum_arrivals)
```

```python
print(f"Average number of cars that arrived in 24 hours: {int(np.sum(tot_arrivals)/N)}")
print(f"Average time when the first car arrived: {np.mean(first_car):.2f}")
```

## Comparison 

```python
# Compare all 4 plots
fig, axs = plt.subplots(1, 2, figsize = [12, 5])

axs[0].plot(x_axis, MEAN_ARRIVALS[0], c= "blue")
axs[0].plot(x_axis, MEAN_ARRIVALS[1], c= "orange")
axs[0].set_xlim(0,T)
axs[0].set_xlabel ('Time in minutes', fontsize = 8)
axs[0].set_ylabel ('Avg number of cars', fontsize = 10)
axs[0].set_title(f'Avg number of cars', fontstyle = 'italic')
axs[0].grid()

axs[1].plot(x_axis, CUM_ARRIVALS[0], c= "blue")
axs[1].plot(x_axis, CUM_ARRIVALS[1], c= "orange")
axs[1].set_xlim(0,T)
axs[1].set_xlabel ('Time in minutes', fontsize = 8)
axs[1].set_ylabel ('Cumulative number of cars', fontsize = 10)
axs[1].set_title(f'Cumulative number of cars', fontstyle = 'italic')
axs[1].grid()
```

By the above plots and numbers it is easy to check that the algorithms simulating the **inhomogeneous Poisson process** work right.
